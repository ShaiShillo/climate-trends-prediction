```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from datetime import date, timedelta
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
def get_weather_data7(year, month, day):
    # Open browser and navigate to website
    browser = webdriver.Chrome()
    browser.get("https://www.wunderground.com/history/daily/il/haifa/LLHA")
    
    #init var
    year = str(str((2023-year))+':'+' '+str(year))
    day = str((day-1))+':'+' '+str(day)
    # Select year, month, and day from the drop-down menus
    year_select = browser.find_element_by_xpath('//select[@name="year"]')
    year_select.find_element_by_xpath(f'option[@value="{year}"]').click()
    month_select = browser.find_element_by_xpath('//select[@name="month"]')
    month_select.find_element_by_xpath(f'option[@value="{month}"]').click()
    day_select = browser.find_element_by_xpath('//select[@name="day"]')
    day_select.find_element_by_xpath(f'option[@value="{day}"]').click()

    # Click on "Go" button
    go_button = browser.find_element_by_xpath('//button[@type="submit"]')
    go_button.click()

    # Wait for the table to load
    wait = WebDriverWait(browser, 10)
    table = wait.until(EC.presence_of_element_located((By.XPATH, '//table[@class="row"]')))

    # Extract the data from the table
    temperature = []
    precipitation = []
    wind = []
    rows = table.find_elements_by_xpath('.//tr')
    for row in rows:
        cells = row.find_elements_by_xpath('.//td')
        temperature.append(cells[1].text)
        precipitation.append(cells[2].text)
        wind.append(cells[3].text)

    # Close the browser
    browser.quit()

    # Return the data as a list
    return [year, month, day, temperature, precipitation, wind]
```

```python
def get_weather_data2(year, month, day):
    # Open browser and navigate to the website
    browser = webdriver.Chrome()
    browser.get('https://www.wunderground.com/history/daily/il/haifa/LLHA')

    # Select year, month, and day from the drop-down menus
    year_select = Select(browser.	find_element(By.ID,'yearSelection'))
    year_select.select_by_value(str(year))
    month_select = Select(browser.find_element(By.ID,'monthSelection'))
    month_select.select_by_value(str(month))
    day_select = Select(browser.find_element(By.ID,'daySelection'))
    day_select.select_by_value(str(day))

    # Click the "View" button to load the data for the selected date
    browser.find_element('dateSubmit').click()

    # Wait for the data to load
    WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.ID, 'dataTable')))

    # Extract the data from the table using BeautifulSoup
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    table = soup.find('table', {'id': 'dataTable'})
    rows = table.find_all('tr')
    temperature = []
    precipitation = []
    wind = []
    for row in rows:
        cells = row.find_all('td')
        temperature.append(cells[1].text)
        precipitation.append(cells[2].text)
        wind.append(cells[3].text)

    # Close the browser
    browser.quit()

    # Return the data as a list
    return [year, month, day, temperature, precipitation, wind]
```

```python
def get_weather_data(year, month, day):
    # Open browser and navigate to the website
    browser = webdriver.Chrome()
    browser.get('https://www.wunderground.com/history/daily/il/haifa/LLHA')

    # Select year, month, and day from the drop-down menus
    year_select = Select(browser.find_element(By.ID,'yearSelection'))
    year_select.select_by_value(str(year)+':')
    month_select = Select(browser.find_element(By.ID,'monthSelection'))
    month_select.select_by_value(str(month))
    day_select = Select(browser.find_element(By.ID,'daySelection'))
    day_select.select_by_value(str(day))

    # Click the "View" button to load the data for the selected date
    browser.find_element(By.ID,'dateSubmit').click()

    # Wait for the data to load
    WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.ID, 'dataTable')))

    # Extract the data from the table using BeautifulSoup
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    table = soup.find('table', {'id': 'dataTable'})
    rows = table.find_all('tr')
    temperature = []
    precipitation = []
    wind = []
    for row in rows:
        cells = row.find_all('td')
        temperature.append(cells[1].text)
        precipitation.append(cells[2].text)
        wind.append(cells[3].text)

    # Close the browser
    browser.quit()

    # Return the data as a list
    return [year, month, day, temperature, precipitation, wind]
```

```python
browser = webdriver.Chrome()
def get_weather_data_for_dates(start_year, start_month, start_day, end_year, end_month, end_day):
    weather_data = {}
    current_date = datetime(start_year, start_month, start_day)
    end_date = datetime(end_year, end_month, end_day)
    while current_date <= end_date:
        try:
            year, month, day = current_date.year, current_date.month, current_date.day
            date_data = get_weather_data3(year, month, day)
            weather_data[current_date.strftime("%Y-%m-%d")] = date_data
        except Exception as e:
            print("An error occurred while getting data for date", current_date, ":", e)
        current_date += timedelta(days=1)
    return weather_data


weather_data = get_weather_data_for_dates(2012, 1, 1, 2012, 1, 6)
weather_data_df = pd.DataFrame.from_dict(weather_data)
browser.quit()
print(weather_data)
```
